```python
def softmax_normalize(x):
    b = x.max()
    y = np.exp(x - b)
    return y / y.sum()

def replay(ltmemory):
    model.printWeightAverages()
    minibatch = random.sample(ltmemory, len(ltmemory))

    training_states = np.array(
        [model.convertToModelInput(row['state']) for row in minibatch])
    training_targets = {'value_head': np.array(
        [row['value'] for row in minibatch]), 'policy_head': np.array([row['AV'] for row in minibatch])}
    print(training_states, training_targets)
    fit = model.fit(training_states, training_targets,
                         epochs=config.TRAINING_LOOPS, verbose=1, validation_split=0, batch_size=32)
    model.printWeightAverages()
    print('NEW LOSS %s', fit.history)
    
    y = model.predict(training_states)
    vh = y[0]
    ph = y[1]
    for i in range(30):
        print(training_states[i])
        print(vh[i], training_targets['value_head'][i])
        print(softmax_normalize(ph[i]), training_targets['policy_head'][i])
        
mem = pickle.load(open("run/memory/memory0007.p", "rb"))
ltmemory = mem.ltmemory
env = Game()                  
model = Densely_connected_net(
    config.REG_CONST, config.LEARNING_RATE, env.input_shape,   env.action_size)
# model.viewLayers()
                  
replay(ltmemory)
```
